````python
import numpy as np
import librosa.display
import IPython.display as ipd
import librosa
````

````python
sample_data = '/content/gdrive/MyDrive/음소인식project/project_dataset/kspon_train/KsponSpeech_01/KsponSpeech_0001/KsponSpeech_000007.pcm'
sample_data
````

````python
with open(sample_data, 'rb') as opened_pcm_file:
  buf = opened_pcm_file.read()
  pcm_data =np.frombuffer(buf, dtype = 'int16')
  wav_data = librosa.util.buf_to_float(pcm_data,2)
signal = np.memmap(sample_data, dtype = 'h', mode = 'r').astype('float32')
signal = signal /32767
ipd.Audio(wav_data, rate=16000)
````
